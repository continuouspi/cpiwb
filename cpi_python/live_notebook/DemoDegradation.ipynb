# Synthesis and Degradation

#### Author(s):  Stanley Wang
#### Reference: [Tyson et al (2003)](http://www.sciencedirect.com/science/article/pii/S0955067403000176) | [Download](https://cpi-catalog.000webhostapp.com/cpi-files/synthesis_and_degradation.cpi)

This is a model of a common dynamic in proteins: the synthesis and degradation of a molecule. It is the 1st model from the Tyson et al (2003) paper.

We have a signal $S$ which reacts to promote the generation of a product $R$: this reaction happens at some unknown rate which we can represent as $k1$ . $R$ is produced by some constant ambient signal which is always in the system and also decays out of the system. The authors provide an example of the species S as the concentration of mRNA while R can be thought of as the concentration of protein - “the mRNA can be used to interlace amino acids in the protein”. We can define the rate of production of $R$ by some ambient signal as $k0$ and the rate of decay of $R$ by $k2$ .
