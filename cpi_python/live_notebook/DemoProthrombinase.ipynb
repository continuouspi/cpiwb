A kinetic model of how thrombin is generated from prothrombin with the intermediates of meizothrombin and prethrombin-2.

Thrombin is an enzyme that plays a role in blood coagulation. It converts soluble fibrinogen into insoluble strands of fibrin, and it also catalises other coagulation-related reactions. Thrombin is generated enzymatically from prothrombin by 2 pathways. There had been conflicting results about the exact reaction rates in the generation of thrombin. The authors of this paper took the conflicting datasets of 2 independent research groups and created a unified kinetic model based on them, then verified the results experimentally. In terms of Cπ code, this is a simple system with 4 reactions, where in each reaction one substance simply changes into another at a certain rate. The plot below shows the reaction rates.
