## Simulate the selected process

### simulate_process(self, solver, is_save)

'solver' determines which solver will solve of this process, you can select from 'odeint' or 'GLIMDA'.

'is_save' determines whether to save the simulation results of this process, you can enter 'Yes' or 'No'.

The default ode solver is 'odeint' and it can solve most of the models. If the result of 'odeint' cannot solve the select process, you can use the 'GLIMDA' solver.

```python
simulate1=cpi_model.simulate_process('odeint', 'No')
```
